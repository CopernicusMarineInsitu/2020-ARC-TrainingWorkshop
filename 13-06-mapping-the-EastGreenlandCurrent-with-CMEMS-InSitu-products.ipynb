## 1. Introduction
[Go back to the "Table of Contents"](#Table-of-Contents)

CMEMS catalog enable users to access a wide range of data products based on model forecasts and observations (remote and insitu). In this notebook we aim to showcase its usability for monitoring the East Greenland Current, one of the major reponsible for Ice transportation (cold and freswater inputs) off the Arctic region and into the adjacent seas (i.e Labrador Sea and North Atlantic Ocean). This current is can be envisaged as the counterpart of the North Atlantic Current (warm and salty), that in this case discurs in the oposite direction (south-nort) along the Norwegian Sea. The implications of the any strengthening of this current are numerous (i.e find some described [here](https://www.nature.com/articles/s41598-019-53723-z)) and therefore its monitoring is key for undertanding upcomming changes in the area. 

| ![EastGreenlandCurrent.jpg](./img/EastGreenlandCurrent.jpg)| 
|:--:| 
| *oceanographic surface currents around Iceland. Image source: http://www.hi.is/~jeir/panis_currents.html* |

On next sections we will focus in the insitu observations that are available for this task trough the Copernicus Marine Service Catalog in this area ([`INSITU_ARC_NRT_OBSERVATIONS_013_031`](https://resources.marine.copernicus.eu/?option=com_csw&view=details&product_id=INSITU_ARC_NRT_OBSERVATIONS_013_031): standarized and quality-checked observations of oceanographic and bgc parameters from a wide range of platforms (drifiters,profilers,moorings etc) and providers. 

## 2. In Situ Observations
[Go back to the "Table of Contents"](#Table-of-Contents)

The In Situ observations are on-site measurements of [physical and biogeochemical](https://archimer.ifremer.fr/doc/00422/53381/) parameters performed by means of a wide range of instruments and/or platforms de ployed in the ocean. To discover which ones are available in the area to monitor the East Greenland Current we will use the [Dashboard](http://www.marineinsitu.eu/dashboard/): an interactive tool that enables to access the data reported by platforms deployed in a given region at some point (still operative or smatled ones).<br> This tool is accessible as `Advanced tool` for visualization in the Copernicus Marine Service Catalog for insitu miltparamter products (see [here](https://view-cmems.mercator-ocean.fr/INSITU_ARC_NRT_OBSERVATIONS_013_031) for the ARCTIC)

*Note: Additional sources (i.e images and numbers) generated by the the insitu notebooks prepared for the [2020 CMEMS ARC Workshop](https://marine.copernicus.eu/wp-content/uploads/2020/11/CMEMS-Arctic-Online-Training-Workshop-Agenda_v5.pdf) will be used as well*.

### 2.1. Profiles

Profile*-like* data provides an overwiew of the water column in terms of a particular property (temperature, salinity etc) and can be obtained from a great variety of instruments/platforms. **Let's get a quick glimpse of the amount of profile-*like* data that is available in the arctic area within the last 30 days**: [**click here**](http://instac.socib.es/dashboard/?mapCenterLat=72.34&mapCenterLng=-30.85&mapZoomLevel=3&feature_like=PR&dataset_like=arc_multiparameter_nrt&archives_like=latest&addonCredits=false)

### 2.2. Trajectories
[Go back to the "Table of Contents"](#Table-of-Contents)

Trajectory-*like* data provides an overwiew of the surface water at a given *transitory* location, in terms of a particular property (temperature, salinity etc) and can be obtained from a great variaty of platforms. **Let's get a quick glimpse of the amount of trajectory-*like* data that is available in the arctic area within the last 30 days**: [**click here**](http://instac.socib.es/dashboard/?mapCenterLat=69.16&mapCenterLng=-5.71&mapZoomLevel=3&dataset_like=arc_multiparameter_nrt&archives_like=latest&type_like=DB&addonCredits=false)

| ![*trajectories-dashboard.png*](./img/trajectories-dashboard.png)| 
|:--:| 
| snapshot of the latest trajectories in the dashboard (Arctic)|

Up to today, the greater and only provider of trajectory data in the arctic region are the Drifting buoys (in green above): mobile platforms that measures a wide range of parameters while driftting with the ocean currents. 

These platforms can be used therefore to characterize its driving current (i.e The East Greenland Current!!) in a number of ways: either by direct measurements of the water they travel with (temperature, salinity) or by delay time analysis of its motion (to derive the velocity).

### 2.4. Grids
[Go back to the "Table of Contents"](#Table-of-Contents)

Grid-like data provides an overwiew of the surface water velocity in a given area over time and its is presently produced by just one type of platform: the High Frequency Radars. **Let's get a quick glimpse of the amount of Grid data that is available in the arctic area within the last 30 days**: [**click here**](http://instac.socib.es/dashboard/?mapCenterLat=65.22&mapCenterLng=6.64&mapZoomLevel=4&type_like=HF&dataset_like=arc_multiparameter_nrt&archives_like=latest&addonCredits=false)

| ![*hfradars-dashboard.png*](./img/hfradars-dashboard.png)| 
|:--:| 
| snapshot of the latest HF Radars in the dashboard (Arctic) |

High Frequency Radars are composed by two or more antenas (dots above) and provides two types of files: 
- Radial Velocity files (RV) produced by a single antenna and accessible by clicking the dot.
- Total Velocity files (TV) a combination of the data coming from more than one antenna and accessible by clicking the area.

*Note: In the above maps, only the coverage displayed corresponds to the last water velocity fields qualified as 'good' (QCflag = 1).*

There are still very few High Frequency Radar platforms in the Arctic but when talking about monitoring current patterns near the coast these cannot be forgotten. Find next how well High Frequency Radars do the job!:
